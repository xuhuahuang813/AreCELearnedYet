```python
import re
trueList = []
estiList = []
QList = []

# 打开文件
with open('output/result/merge-1w-dynamic/mscn.txt', 'r') as file:
    # 定义正则表达式模式
    pattern = r'【(true|esti|Q)】(\d+\.?\d*)'

    # 逐行读取文件内容
    for line in file:
        # 在每一行中搜索匹配的模式
        matches = re.findall(pattern, line)
        
        # 初始化字典来存储匹配到的内容
        match_dict = {'true': None, 'esti': None, 'Q': None}
        for match in matches:
            label = match[0]
            value = match[1]
            match_dict[label] = value
        
        # 检查是否每个标签都有值，如果有则打印出来
        if all(match_dict.values()):
            # print("true:", match_dict['true'])
            # print("esti:", match_dict['esti'])
            # print("Q:", match_dict['Q'])
            trueList.append(float(match_dict['true']))
            estiList.append(float(match_dict['esti']))
            QList.append(float(match_dict['Q']))
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['WenQuanYi Micro Hei']
plt.figure(figsize=(4, 4))  
# 绘制散点图
plt.scatter(trueList, estiList, c='k', marker='o', s=3)
plt.plot([0, 70000], [0, 70000], color='red')
plt.fill_between([0, 70000], [0.9*0, 1.1*70000], [1.1*0, 0.9*70000], color='gray', alpha=0.3)
# 添加标题和标签
# plt.title('散点图示例')
plt.xlabel('真实基数（万）', fontsize=20)
plt.ylabel('预估基数（万）', fontsize=20)

# 设置x轴和y轴的刻度
plt.xticks([10000, 30000, 50000, 70000], ['1', '3', '5', '7'], fontsize=20)
plt.yticks([10000, 30000, 50000, 70000], ['1', '3', '5', '7'], fontsize=20)

plt.savefig('./output/result/merge-1w-dynamic/mscn.png', dpi=500)
# plt.show()
```
