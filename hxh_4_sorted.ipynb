```python
import pandas as pd
import time

# 记录代码开始时间（毫秒级别）
start_time = time.time() * 1000

selected_columns = ['education', 'education_num', 'occupation']
selected_data = data.data[selected_columns].reset_index(drop=True)

# Sort by 'education', 'education_num', and 'occupation' in ascending order
df_sorted = selected_data.sort_values(by=selected_columns).reset_index(drop=True)

# 使用 value_counts 统计每个组合的数量
count_series = df_sorted.groupby(selected_columns).size().reset_index(name='count')

# 计算概率和累积概率
count_series['Pro'] = count_series['count'] / len(df_sorted)
count_series['CPro'] = count_series['Pro'].cumsum()

# 保留 count 大于 0 的数据，并保持顺序
filtered_data = pd.merge(df_sorted, count_series, on=selected_columns, how='inner')
filtered_data = filtered_data[filtered_data['count'] > 0]

# 保留相同的selected_columns只保留一行
filtered_data = filtered_data.drop_duplicates(subset=selected_columns)

# 记录代码结束时间（毫秒级别）
end_time = time.time() * 1000

# 计算执行时间
execution_time = end_time - start_time

# 显示执行时间
print(f"代码执行时间：{execution_time:.2f} 毫秒")
# 打印结果
print(filtered_data)
```

```python
import pandas as pd
import time

# 记录代码开始时间（毫秒级别）
start_time = time.time() * 1000

selected_columns = ['education', 'education_num', 'occupation']
selected_data = data.data[selected_columns].reset_index(drop=True)

# Sort by 'education', 'education_num', and 'occupation' in ascending order
df_sorted = selected_data.sort_values(by=selected_columns).reset_index(drop=True)

# 使用 value_counts 统计每个组合的数量
count_series = df_sorted.groupby(selected_columns).size().reset_index(name='count')
count_series = count_series[count_series['count'] > 0]

# Calculate the cumulative probability directly using cumsum on the size
cumulative_probability = count_series['count'].cumsum() / len(df_sorted)

# Combine cumulative_probability with the original count_series
result_df = pd.concat([count_series, cumulative_probability], axis=1, ignore_index=True)
result_df.columns = list(count_series.columns) + ['cumulative_probability']

# 记录代码结束时间（毫秒级别）
end_time = time.time() * 1000

# 计算执行时间
execution_time = end_time - start_time

# 显示执行时间
print(f"代码执行时间：{execution_time:.2f} 毫秒")

# Display the result
print(result_df)
```
